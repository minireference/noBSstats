### E1.7

Open the data file [`datasets/players.csv`](http://nobsstats.com/datasets/players.csv)
using [LibreOffice](https://www.libreoffice.org/)
or another spreadsheet program.
Compute the mean and the standard deviation of the variable `age` using spreadsheet functions.

Hint: Use the spreadsheet functions `AVERAGE()` and `STDEV()`.

Assuming you used the default options when loading the data file `players.csv`,
the `age` variable will be the column `C` of the spreadsheet.
You can then compute the average age, by entering the formula `=AVERAGE(C2:C13)`
in a new spreadsheet cell.

Here is the Pandas equivalent of the same operation.

```python
import pandas as pd
players = pd.read_csv("datasets/players.csv")
players["age"].mean()
```

To get the standard deviation the age variable,
we need to use the spreadsheet formula `=STDEV(C2:C13)`.

Here is the Pandas equivalent,
using the `.std()` method.

```python
players["age"].std()
```

### E1.15

Load the kombucha dataset from `datasets/kombucha.csv` 
and extract the kombucha volume measurements from Batch 10,
storing the result in a new variable called `ksample10`.

- **a)** Generate a strip plot of `ksample10`.
- **b)** Compute the mean and the standard deviation of `ksample10`.
- **c)** Select the subset of values in `ksample10` that are less than 1050 and store your result as `ksample10_clean`.
- **d)** Compute the mean and the standard deviation of `ksample10_clean`.


```python
import pandas as pd
kombucha = pd.read_csv("datasets/kombucha.csv")
batch10 = kombucha[kombucha["batch"]==10]
ksample10 = batch10["volume"]

# a)
import seaborn as sns
sns.stripplot(x=ksample10);
```

```python
# b)
ksample10.mean(), ksample10.std()
```

```python
# c)
ksample10_clean = ksample10[ksample10 < 1050]
```

```python
# d)
ksample10_clean.mean(), ksample10_clean.std()
```
