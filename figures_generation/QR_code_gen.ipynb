```python
import qrcode
```

```python
tutorials = [
    "https://minireference.com/static/tutorials/sympy_tutorial.pdf",
    "https://minireference.com/static/tutorials/linear_algebra_in_4_pages.pdf",
    "https://minireference.com/static/tutorials/mech_in_7_pages.pdf",
    "https://minireference.com/static/conceptmaps/math_and_physics_concepts.pdf",
    "https://minireference.com/static/conceptmaps/linear_algebra_concepts.pdf",
    "https://minireference.com/static/conceptmaps/math_concepts.pdf",
    "https://minireference.com/static/conceptmaps/statistics_concepts.pdf",
    "https://raw.githubusercontent.com/minireference/calculus_tutorial/refs/heads/main/calculus_tutorial.pdf",
]

for tutorial in tutorials:
    part = tutorial.rstrip(".pdf")
    name = part.split('/')[-1]
    print("Processing", name, tutorial)
    filename = name + "_QR_code.png"
    img = qrcode.make(tutorial, border=1)
    img.save("figures/QRcodes/" + filename)

print("See figures/QRcodes/")
```

```python

books = [
    "https://minireference.com/static/excerpts/noBSmathphys_v5_preview.pdf",
    "https://minireference.com/static/excerpts/noBSLA_v2_preview.pdf",
    "https://minireference.com/static/excerpts/noBSmath_v5_preview.pdf",
    "https://minireference.com/static/excerpts/noBSstats/",
]


for book in books:
    part = book.rstrip(".pdf/")
    name = part.split('/')[-1]
    print("Processing", name, book)
    filename = name + "_QR_code.png"
    img = qrcode.make(book, border=1)
    img.save("figures/QRcodes/" + filename)

print("See figures/QRcodes/")
```
