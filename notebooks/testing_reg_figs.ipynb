```python
import pandas as pd
```

```python
students = pd.read_csv("../datasets/students.csv")
doctors = pd.read_csv("../datasets/doctors.csv")
```

```python
from ministats import fit_and_plot_reg


efforts = students["effort"]
scores = students["score"]

fit_and_plot_reg(efforts, scores)
```

```python
# FIGURES ONLY
from ministats import plot_reg
from ministats import plot_pred_bands
import statsmodels.formula.api as smf

lm1 = smf.ols("score ~ 1 + effort", data=students).fit()
plot_reg(lm1);
plot_pred_bands(lm1, ci_mean=True, ci_obs=True)
```

```python
from ministats import plot_resid

plot_resid(lm1, lowess=True);
```

```python
lm2 = smf.ols("score ~ 1 + alc + weed + exrc", data=doctors).fit()

from ministats import plot_partreg
plot_partreg(lm2,  "exrc");
```

```python
plot_resid(lm2, lowess=True);
```
