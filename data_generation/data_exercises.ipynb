```python
import random
import pandas as pd
import numpy as np
from scipy.stats.distributions import norm
```

### Data melt operation

```python
random.seed(16)

nstudents = 7

grades_data = {
    "student_ID": 100 + np.arange(1, nstudents+1),
    "test1": norm.rvs(60, 10, size=nstudents).round(0),
    "test2": norm.rvs(70, 10, size=nstudents).round(0),
    "test3": norm.rvs(80, 10, size=nstudents).round(0),
}

df1 = pd.DataFrame(grades_data, dtype="int")
df1
```
