```python
import random
import pandas as pd
import numpy as np
from scipy.stats.distributions import norm
```

### Data melt operation

```python
random.seed(16)

nstudents = 7

grades_data = {
    "student_ID": (100 + np.arange(1, nstudents+1)).astype(int),
    "name": ["Amelie", "Franklin", "Marge", "Gab", "Caroline", "Nora", "Saul"],
    "test1": norm.rvs(60, 10, size=nstudents).round(0).astype(int),
    "test2": norm.rvs(70, 10, size=nstudents).round(0).astype(int),
    "test3": norm.rvs(80, 10, size=nstudents).round(0).astype(int),
}

markswide = pd.DataFrame(grades_data)
markswide
```

```python
markswide.to_csv("../datasets/markswide.csv", index=False)
```

```python
markswide2 = pd.read_csv("../datasets/markswide.csv")
markswide2.equals(markswide)
```

```python
markswide.melt(id_vars=["student_ID", "name"], var_name="test", value_name="grade") \
         .replace({"test":{"test1":1, "test2":2, "test3":3}})
```
